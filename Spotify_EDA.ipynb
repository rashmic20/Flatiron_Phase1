```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

spotify = pd.read_csv("spotify-2023.csv", encoding="latin-1")
```

```python
spotify.head()
```

```python
spotify.nunique()
```

```python
spotify.info()
```

```python
# No. of songs released each year

Released_year = spotify.released_year.value_counts().reset_index()
```

```python
Released_year.sort_values(by='released_year',ascending=False)
```

```python
sns.countplot(data=spotify,x="released_month");
```

```python
# Artist with the most release
Artist_Most_release = spotify['artist(s)_name'].value_counts()
```

```python
Artist_Most_release.head(1)
```

```python
# Top 10 Artist with most release
Top10_Artist = Artist_Most_release.head(10)
```

```python
Top10_Artist
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=Top10_Artist.values,y=Top10_Artist.index,palette='viridis')
plt.xlabel('No. of songs')
plt.ylabel('Artist(s)')
plt.title('Top 10 artist with most number of songs')
plt.show;
```
